```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from itertools import combinations
```

```python
G = nx.Graph()
G.add_edges_from(list(nx.fast_gnp_random_graph(7, 0.7).edges))
G.add_edges_from([(a+7, b+7) for a,b in list(nx.fast_gnp_random_graph(7, 0.7).edges)])
G.add_edge(4,12)               
```

```python
colors = ['#2A9D8F','#2A9D8F','#2A9D8F','#2A9D8F','#2A9D8F','#2A9D8F','#2A9D8F','#E76F51','#E76F51','#E76F51','#E76F51','#E76F51','#E76F51','#E76F51']
fig = plt.figure(figsize=(4,4))
ax = plt.axis()
nx.draw_kamada_kawai(G, node_color=colors, edge_color='gray', node_size=100)
plt.savefig("easy_clustering.eps", dpi=300, bbox_inches = 'tight', pad_inches = 0.1)
```
